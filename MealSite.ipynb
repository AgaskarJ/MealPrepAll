```python
import requests
from bs4 import BeautifulSoup

```

```python
# Specify url: url
url = 'https://www.budgetbytes.com/spaghetti-with-vegetable-meat-sauce/'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extracts the response as html: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(r.content, 'html.parser')#html_doc

# Prettify the BeautifulSoup object: pretty_soup
pretty_soup = soup.prettify()

# Print the response
print(pretty_soup)
```

```python
# Get the title of Guido's webpage: guido_title
guido_title = soup.title

# Print the title of Guido's webpage to the shell
print(guido_title.string)
```

```python
ingredient2 = []
for ingredient in soup.find_all('div', {'class': 'wprm-recipe-ingredient-group'}):
    ing_name = ingredient.find('ul', {'class': 'wrpm-recipe-ingredients'}).string
    ingredient2.append(ing_name)

print(ingredient2)
```

```python
# Find all 'a' tags (which define hyperlinks): a_tags
div_tags = soup.find(class_='wprm-recipe-ingredients')
#print(div_tags)

lis = div_tags.contents
##print(lis[0].next_sibling)

ingredient = []
for li in lis: 
    for rec in li.contents[:5]: 
        item = rec.string
        ingredient.append(item)

while (' ' in ingredient):
    ingredient.remove(' ')

name = [*enumerate(ingredient,1)]

ing_name = ''
for x in ingredient[:3]: 
    ing_name = ' '.join(ingredient)

print(ing_name)
        
# Print the URLs to the shell
#for link in a_tags:
   # print(link.get('href'))
```

```python
new = []
for name in soup.find_all("li", {"class": "wprm-recipe-ingredient"}):
    amt = name.find("span", {"class": "wprm-recipe-ingredient-amount"}).text
    #unit = name.find("span", {"class": "wprm-recipe-ingredient-unit"}).content
    names = name.find("span", {"class": "wprm-recipe-ingredient-name"}).get_text()
    final_name = ' '.join((amt,names))
    new.append(final_name)

print(new)
```
